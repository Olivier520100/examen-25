```python
dfproducts = pd.read_csv("products.csv", sep=",")
dforders = pd.read_csv("orders.csv", sep=",")
dfcustomers = pd.read_csv("customers.csv", sep=",")


```

Creates combo table list of the most popular combinations

```python
combinationtable = np.zeros((len(dfproducts),len(dfproducts)))

orderpercustomerid = {}
print((len(dfproducts)))
print(combinationtable.shape)
combolist = []
currentnumset = set()
prevorderid = ""
for index,row in dforders.iterrows():
  currentorderid = row["OrderID"]
  if currentorderid != prevorderid:
    prevorderid = currentorderid
    combolist.append(currentnumset)
    currentnumset = set()
  currentnumset.add(int(row["ProductID"][1:]))

for currentset in combolist:
  currentlist = list(currentset)
  i = 0
  j = 0
  while i < len(currentlist):
    j=i+1
    while j < len(currentlist):
      combinationtable[currentlist[i],currentlist[j]]+=1
      combinationtable[currentlist[j],currentlist[i]]+=1

      j+=1
    i+=1

print(combinationtable)



```
